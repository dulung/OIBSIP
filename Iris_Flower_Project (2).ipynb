##Data Visualization and Analysis

*We are importing seaborn and mathplotlib libraries and then computing the corelation matrix and plotting the heatmap to understand the correlation between each of them*

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Compute the correlation matrix
corr_matrix = df.drop(['Id','Species'],axis=1).corr()
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

##Data Preparation

*We convert the Species column contents into numerical values*

```python
df['Species'] = df['Species'].map({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
df
```

*We are identifying the dependent and indepedent columns and storing the dependent values in y and the independent values in X*

```python
X=df.drop(['Id','Species'],axis=1)
y=df['Species']
```

*We are dividing the data into training and testing sets*

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

##Model Builiding and Evaluation

*For Model Building I have used the algorithm Logistic Regression*

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(max_iter=500)
log_reg.fit(X_train,y_train)
print("Accuracy:",log_reg.score(X_test,y_test))
```
