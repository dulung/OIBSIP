#**CAR PRICE PREDICTION USING MACHINE LEARNING**

## Data Collection

*We upload the csv file*

*We import pandas library*

```python
import pandas as pd
```

*We read the csv file and store the data in a dataframe*

```python
df=pd.read_csv('car data.csv')
df
```

##Data Cleaning

*We are checking if there is any null value present in the dataset*

```python
df.isnull().sum()
```

##Data Visualization and Analysis

*We are visualizing the data using pairplot*

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df,x_vars=['Year', 'Present_Price', 'Driven_kms', 'Fuel_Type', 'Selling_type', 'Transmission', 'Owner'], y_vars=['Selling_Price'], height=5, aspect=1, kind='scatter')
plt.show()
```

##Data Preperation

*We convert the column contents into numerical values*

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Fuel_Type'] = label_encoder.fit_transform(df['Fuel_Type'])
df['Selling_type'] = label_encoder.fit_transform(df['Selling_type'])
df['Transmission'] = label_encoder.fit_transform(df['Transmission'])
df
```

*We are identifying the dependent and indepedent columns and storing the dependent values in y and the independent values in X*

```python
X = df[['Year', 'Present_Price', 'Driven_kms', 'Fuel_Type', 'Selling_type', 'Transmission', 'Owner']]
y = df['Selling_Price']
```

*We are dividing the data into training and testing sets*

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=142)
```

##Model Builiding and Evaluation

*For Model Building, I have used the algorithm Gradient Boosting Regressor*

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(n_estimators=100)
model.fit(X_train, y_train)
print("Accuracy: ", model.score(X_test,y_test))
```
